# **Deteccin de objetos aplicada al conteo de vacunos**
## **Object detection for cattle counting**
John Fredy Ramirez Agudelo, Zoot, MSc, PhD; Sebastin Bedoya Mazo, Zoot; Jos Fernando Guarn Montoya, Zoot, MSc, PhD

Grupo de Investigacin en Ciencias Agrarias (GRICA), Facultad de Ciencias Agrarias, Universidad de Antioquia, Medelln, Colombia. E-mail: precisionlivestockfarming@gmail.com


**Antecedentes**: el conteo de animales es una prctica rutinaria en muchos puntos de la cadena productiva vacuna, y gracias a los avances en Inteligencia Artificial, es posible automatizar esta actividad mediante la utilizacin de la deteccin de objetos. Sin embargo, los modelos de deteccin de objetos, como YOLO v4, no pueden ser desplegados en dispositivos mviles o computadores de bajo rendimiento debido a la alta demanda de recursos computacionales de estos modelos. Como alternativa, modelos ms pequeos, como YOLOv4-tiny, puede ser utilizado, ya que estos modelos enfatizan en la velocidad ms que en la precisin de las predicciones, lo cual resulta muy til para entornos informticos limitados como telfonos mviles. **Objetivo**: Desarrollar una aplicacin para Android que permita realizar conteos rpidos de animales con el modelo YOLOv4-tiny. **Mtodos**: 4000 fotos, con al menos un vacuno por foto, fueron etiquetadas con el programa LabelImg, el etiquetado de imgenes es el proceso de identificar y seleccionar manualmente las reas donde se ubican los objetos de inters en una imagen. Los recursos gratuitos en Google Colab y el cdigo abierto de Darknet fueron utilizados para re-entrenar el modelo YOLOv4-tiny con dicha base de datos. Despus del proceso de entrenamiento, el modelo fue desplegado en una aplicacin para Android con el programa Android Studio. **Resultados**: El re-entrenamiento del modelo YOLOv4-tiny, para identificar solo un tipo de objeto (vacunos), incremento significativamente la precisin de este modelo para dicha tarea. La utilizacin de la aplicacin Android en distintos entornos (potreros, corrales, salas de ordeo, etc) permiti reconocer las ventajas de este tipo de herramientas. **Conclusiones**: los resultados obtenidos demuestran que la utilizacin de modelos pequeos de Machine Learning, entrenados para tareas especficas, mejora sustancialmente el desempeo de estos modelos y permite el desarrollo de herramientas tecnolgicas para la implementacin de la Ganadera de Precisin a un bajo costo.

**Palabras clave**: ganadera de precisin, deteccin de objetos, YOLOv4-tiny.

# **Visin artificial aplicada al monitoreo de vacunos: Deteccin de posicin** #
## **Artificial vision applied to cattle monitoring: Pose detection**
John Fredy Ramirez Agudelo, Zoot, MSc, PhD; Sebastin Bedoya Mazo, Zoot; Jos Fernando Guarn Montoya, Zoot, MSc, PhD

Grupo de Investigacin en Ciencias Agrarias (GRICA), Facultad de Ciencias Agrarias, Universidad de Antioquia, Medelln, Colombia. E-mail: precisionlivestockfarming@gmail.com


**Antecedentes**: la deteccin de posicin es una tcnica de visin artificial utilizada para determinar cambios espacio-temporales en personas, animales o mquinas articuladas. Teniendo en cuenta que muchas anomalas en el ganado pueden ser detectadas mediante el anlisis de este tipo de cambios (e.g. cojeras, reduccin en el consumo, celo), es posible considerar la utilizacin de esta tcnica para el mejoramiento de los programas de monitoreo individualizado de los animales en las fincas lecheras. Objetivo: explorar la implementacin de la tcnica de deteccin de posicin en vacas lecheras. **Mtodos**: 25 vacas lecheras de la hacienda La Montaa UdeA, fueron grabadas en video (cortos de max.. 5 s) a la salida del ordeo. Los videos fueron procesados con DeepLabCut, que es un paquete en lenguaje Python, de cdigo abierto, rpido y robusto, que se puede utilizar para calcular estimaciones de pose en 3D. Se realizaron tres pasos para el entrenamiento del sistema de deteccin: 1) extraccin de fotogramas, se utilizaron 10 fotogramas por video, 2) etiquetado de los puntos de referencia en cada fotograma, se utilizaron 8 puntos en vista lateral (boca, cuello, escapula, radio-metacarpo, mano, Ileon, tibia-tarso, pata), 3) entrenamiento del modelo, se utiliz el modelo de clasificacin de imgenes Resnet50. El modelo re-entrenado, con 350000 ciclos de iteracin (epochs), fue utilizado para realizar la deteccin de posicin en los videos originales. Finalmente se realiz inspeccin visual del desempeo del modelo. **Resultados**: La inspeccin visual de los videos permiti determinar una alta tasa de eficiencia del modelo en la determinacin de los cambios de posicin en los videos originales validando esta tcnica para el anlisis automatizado de posicin en vacas lecheras ([ver video La Montaa](https://drive.google.com/file/d/1KHH-rxo8tr3RRLJlit8su0L7dBylKD5z/view?usp=sharing)). **Conclusiones**: los resultados obtenidos demuestran que la deteccin automtica de posiciones es una tcnica que puede ser implementada fcilmente para la automatizacin del monitoreo de vacunos y que se pueden conducir a travs de esta ensayos de deteccin de afectaciones a la salud de los animales por cojeras, cambios en el comportamiento de los animales (celos, enfermedad, estrs por calor, etc) y dems aplicaciones que son altamente demandantes de mano de obra calificada y que, con el uso de esta tcnica, son susceptibles de automatizacin. 

**Palabras clave**: DeepLabCut, clasificacin de imagen. 
**Key words**: DeepLabCut, image classification.

# **Clasificacin automtica de andares equinos** #
## **Automatic classification of horse footsteps sound**
John Fredy Ramirez Agudelo, Zoot, MSc, PhD; Sebastin Bedoya Mazo, Zoot; Jos Fernando Guarn Montoya, Zoot, MSc, PhD

Grupo de Investigacin en Ciencias Agrarias (GRICA), Facultad de Ciencias Agrarias, Universidad de Antioquia, Medelln, Colombia. E-mail: precisionlivestockfarming@gmail.com


**Antecedentes**: Los sonidos emitidos por los animales son una fuente de informacin que permite identificar fenotipos, condiciones de estrs, estados de salud, y otros. La escucha y anlisis constante de dichos sonidos solamente puede ser garantizada por sistemas de monitoreo automtico que registren y analicen perturbaciones sonoras, ayudando en la identificacin temprana de problemas productivos. El anlisis de sonidos a travs de imgenes es una tcnica de Inteligencia Artificial til para tal fin. **Objetivo**: evaluar el desempeo de un modelo de clasificacin de imgenes re-entrenado para la identificacin de andares equinos. **Mtodos**: 1000 imgenes de espectrogramas de igual nmero de segundos de grabacin de cuatro andares equinos (Paso Fino Colombiano, Galope, Trocha y Trote) fueron utilizadas para re-entrenar el modelo de clasificacin de imgenes Resnet34 en las cuatro categoras de andares. Las imgenes fueron aleatoriamente divididas en dos sub-grupos para el entrenamiento y validacin del modelo (80, 20 %, respectivamente). La extraccin de audios, su trasformacin en imgenes de espectrograma, el entrenamiento y validacin fueron realizados en Python. **Resultados**: Despus de 50 ciclos de iteracin (epochs) se obtuvo un 93.6, 76, 86.1 y 97.4 % de precisin para la deteccin de los andares Paso Fino Colombiano, Galope, Trocha y Trote, respectivamente. **Conclusiones**: los resultados obtenidos demuestran que la clasificacin de imgenes es una tcnica til para el anlisis de sonidos, que, al ser transformados en imagen, incrementan la facilidad y el volumen de datos para el anlisis y optimizan el uso de recursos computacionales y la capacidad de anlisis con potenciales aplicaciones en salud animal, bienestar animal, conservacin de especies, nutricin animal, entre otras reas de inters veterinario y zootcnico.

# **Utilizacin de dispositivos Android para el monitoreo de la actividad de vacunos en pastoreo** #
## **Grazing-cattle activities recognition with Android devices**
John Fredy Ramirez Agudelo, Zoot, MSc, PhD; Sebastin Bedoya Mazo, Zoot; Jos Fernando Guarn Montoya, Zoot, MSc, PhD; Sandra Lucia Posada Ochoa, Zoot, MSc, DrSc; Jaime Ricardo Rosero Noguera, Zoot, MSc, DrSc

Grupo de Investigacin en Ciencias Agrarias (GRICA), Facultad de Ciencias Agrarias, Universidad de Antioquia, Medelln, Colombia. E-mail: precisionlivestockfarming@gmail.com

**Antecedentes**: el grupo de investigacin GRICA UdeA desarroll una app para dispositivos Android que permite monitorear el patrn diario de tres actividades (consumo, descanso, rumia) en vacunos en pastoreo. Esta tecnologa se basa en la utilizacin de telfonos celulares que registran y analizan los movimientos y posicin de la cabeza del animal. La app utiliza un modelo de red neuronal entrenado para interpretar los cambios de aceleracin registrados por el sensor del dispositivo Android, y de esta manera, determinar la actividad del animal minuto a minuto. **Objetivo**: evaluar la utilizacin de esta tecnologa en el monitoreo de la actividad de vacas lecheras en pastoreo. **Mtodos**: tres telfonos celulares (LG gm730) fueron utilizados para registrar la actividad de n vacas Holstein. Los telfonos fueron posicionados en la parte trasera de la cabeza da las vacas mediante un arns, y fueron retirados dos das despus. Se realizaron observaciones directas y no sistemticas de los animales para contrastar las actividades registradas por el dispositivo. Adicionalmente, se determin la duracin de la batera. **Resultados**: Los tiempos promedio de consumo, rumia y descanso registrados por la app fueron 459, 466 y 505 minutos por da, respectivamente. Se observ coherencia entre las observaciones directas y las registradas por el dispositivo. El dispositivo permiti identificar una alta actividad ingestiva despus de cada ordeo (6h y 15h), actividad que decay al medioda y a las 18h. Actividad nocturna similar para rumia y descanso. Duracin media de la batera de 30 horas. **Conclusin**: los resultados indican que la app desarrollada por el grupo GRICA es una herramienta til para el registro de la actividad individual de vacas lecheras en pastoreo, durante periodos cortos de tiempo (e.g. 24 h).
**Palabras clave**: Ganadera de precisin, actividad en pastoreo, Android. 
**Key words**: Precision Livestock Farming, grazing activity, Android.
